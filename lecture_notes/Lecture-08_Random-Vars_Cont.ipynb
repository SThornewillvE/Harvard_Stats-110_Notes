## Part 1. CDF and PMF

CDF = "Cumulative Distribution Function"

i.e. for a distribution, we add all the values that happened to that point.

$$F(x) = P(X \leq x)$$

We can think of the probability mass function that we had from last lecture of our 10 fair coin flips creating a binomical distribution

```python
# Probability of 1
n = 20  # 10 coin flips
p = 0.5  # Fair coin

# Create x values
x_vals = np.array([binomial(n, p, i) for i in range(0, n+1)])

plt.bar(range(0,n+1), x_vals, alpha=0.6)
plt.xticks(range(0, n+1))
plt.ylabel("Probability")
plt.title("Bernouli Distribution, n={} p={}".format(n, p))
plt.show()
```

We can use the cumulative sum function to change from PMF to CDF

```python
x_cumsum = np.cumsum(x_vals)
```

```python
plt.bar(range(0,n+1), x_cumsum, alpha=0.6)
plt.xticks(range(0, n+1))
plt.ylabel("Probability")
plt.title("Bernouli Distribution (CDF), n={} p={}".format(n, p))
plt.show()
```

Looking at the CDF should remind you of simply integrating the PMF, because essentially that is what is happening.

## Part 2. Hypergeometric distributions

Consider the case where you have a jar of black and white marbles and you want to draw `k` of them. What is the probability of drawing any number of them?

The total number of configurations is:

$$total\ configs = {{w+b} \choose n}$$

And the number of ways that we can draw is

$$draws = {w \choose k}{b \choose {n-k}}$$

Therefore:

$$P(X = k) = \frac{{w \choose k}{b \choose {n-k}}}{{{w+b} \choose n}}$$

```python
def hypergeometic(w, b, n, k):
    return nCr(w, k)*nCr(b, (n-k))/nCr((w+b), n)
```

So let's have a look and plot it.

Say that we have 13 white and 7 black marbles. We want to draw 6 of them.

What does the PMF look like?

```python
w = 13
b = 7
n = 6  # Note that n cannot be larger than either w or b


# Create x valueys
x_vals = np.array([hypergeometic(w, b, n, i) for i in range(0, n+1)])

plt.bar(range(0,n+1), x_vals, alpha=0.6)
plt.xticks(range(0, n+1))
plt.ylabel("Probability")
plt.title("Hypergeometric Distribution, n={} p={}".format(n, p))
plt.show()
```

We can note that if we are drawing from a large pool then the PMF will resemble a binomial distribution because a draw from a large pool will not effect the likelihood of the second one by a lot.
